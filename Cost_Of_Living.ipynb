SOUTH AFRICA DATASET

```python
#create a subset of the DataFrame which has South African Data
df_mzansi=df[df['country']=='South Africa']
df_mzansi
```

```python
#drop the 'country' column in df_mzansi since all the cities are South African
df_mzansi=df_mzansi.drop('country',axis=1)
df_mzansi
```

```python
#set the city column as the index of the dataset
df_mzansi=df_mzansi.set_index('city')
df_mzansi.head(3)
```

At the time (2023/01/21) of this analysis, one 1 United States Dollar is ~ 17 South African Rand


```python
#convert all the values in df_mzansi from USD to ZAR
df_mzansi.loc[:,'x1':'x54']=df_mzansi.loc[:,'x1':'x54'].apply(lambda p:p*17)
df_mzansi
```

```python
#check the number of missing values in each column
df_mzansi_null=df_mzansi.isnull().sum().reset_index().rename(columns={'index':'column',0:'Missing_Values'})
#get the percentage(%) of missing values in each column
df_mzansi_null['Missing_%']=df_mzansi_null['Missing_Values'].apply(lambda n:round((n/df_mzansi.shape[0])*100),2)
df_mzansi_null
```

Some columns appear to have significantly more missing values, but by looking at the dataset it appears to be more row specific in that some rows just have more missing values in most columns

```python
for i in df_mzansi.index:
    print(f" {i} has missing {df_mzansi.loc[i].isnull().sum()} values")
    print(f" And that's {round((((df_mzansi.loc[i].isnull().sum())/(len(df_mzansi.columns)))*100),2)}% of the values missing in the row",end='\n\n')
```

Looks like some cities have a significant amount of missing values so, I'm going to drop cities(rows) that have 10 or more values missing.

```python
before_drop=len(df_mzansi.index)
before_drop
```

```python
for i in df_mzansi.index:
    if (df_mzansi.loc[i].isnull().sum())>=10:
        df_mzansi.drop(i,inplace=True)
```

```python
after_drop=len(df_mzansi.index)
after_drop
```

```python
cities_dropped=before_drop-after_drop
print(f'{cities_dropped} cities didn\'t make the cut')
```

```python
df_mzansi
```

The column 'data_quality' indicates whether or not more comtributors are needed to increase data quality 0 if Numbeo considers that more contributors are needed to increase data quality, else 1, most of the south african cities are 0 so I'm just going to drop the column. also drop the x55 column

```python
df_mzansi=df_mzansi.drop(['data_quality','x55'],axis=1)
df_mzansi.head(3)
```

Create a column that sums up the values in each city to get the cumulative cost of living in each city.

```python
#sum all the values in the columns per row i.e per city to find the total
df_mzansi['Total']=df_mzansi.sum(axis=1)
df_mzansi.head(3)
```

SOUTH AFRICAN CITIES AGRICULTURAL GOOD DATASET

My training is in agricultural economics so I would like to select only a subset of the df_mzansi dataset that includes agricultural/food products.

```python
df_agric_mzansi=df_mzansi.loc[:,'x1':'x27']
df_agric_mzansi
```

```python
#create a 'Groceries' column which is the summation of all the products in df_agric_mzansi dataframe for all the cities.
df_agric_mzansi['Groceries']=df_agric_mzansi.sum(axis=1)
df_agric_mzansi.head(3)
```

```python
df_agric_mzansi[['Groceries']].describe()
```

```python
print(f"The average cost of groceries in South African cities is R{round(df_agric_mzansi['Groceries'].mean(),0)}")
```

```python
most_expensive_groceries=df_agric_mzansi.index[df_agric_mzansi['Groceries']==df_agric_mzansi['Groceries'].max()][0]
cheapest_groceries=df_agric_mzansi.index[df_agric_mzansi['Groceries']==df_agric_mzansi['Groceries'].min()][0]
```

```python
print(f"The city with the most expensive basket of groceries is {most_expensive_groceries}, at R{round(df_agric_mzansi['Groceries'].max(),0)}", end='\n\n')
print(f"The city with the cheapest basket of groceries is {cheapest_groceries},at R{round(df_agric_mzansi['Groceries'].min(),0)}")
```

Let's see where groceries cost more between Johannesburg and Cape Town

```python
groceries_cpt=df_agric_mzansi.loc['Cape Town','Groceries']
groceries_jhb=df_agric_mzansi.loc['Johannesburg','Groceries']
```

```python
if groceries_cpt>groceries_jhb:
    print(f"Groceries cost R{round(groceries_cpt-groceries_jhb,0)} more at Cape Town than Johannesburg")
elif groceries_jhb>groceries_cpt:
    print(f"Groceries cost R{round(groceries_jhb-groceries_cpt,0)} more at Johannesburg than Cape Town")
else:
    print('Groceries cost the same in Johannesburg and Cape Town')
    
    
```
